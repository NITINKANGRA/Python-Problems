2. # **Python program to convert the temperature in degree centigrade to Fahrenheit**

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

celsius = float(input("Enter temperature in degrees Celsius: "))

fahrenheit = celsius_to_fahrenheit(celsius)

print(celsius, "degrees Celsius is equal to", fahrenheit, "degrees Fahrenheit.")
```

9. # **Python program to find the average of 10 numbers using while loop**


```python
# Initialize variables
count = 0
total = 0

# Read 10 numbers from the user
while count < 10:
    num = float(input("Enter a number: "))
    total += num
    count += 1

# Calculate the average
average = total / 10

# Print the result
print("The average of the 10 numbers is:", average)
```

12. # **Python program to find the sum of the digits of an integer using while loop**


```python

num = int(input("Enter an integer: "))

# Initialize variables
sum_of_digits = 0
# the sum of digits
while num > 0:
    digit = num % 10
    sum_of_digits += digit
    num = num // 10

# Print the result
print("The sum of the digits is:", sum_of_digits)
```

14. # **Python program to display all integers within the range 100-200 whose sum of digits is an even number**


```python
# Define the range
start = 100
end = 200

# Iterate through the range
for num in range(start, end + 1):
    # Calculate the sum of digits
    sum_of_digits = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        sum_of_digits += digit
        temp = temp // 10

    # Check if the sum of digits is even
    if sum_of_digits % 2 == 0:
        print(num)
```

15. # **Python program to check whether the given integer is a prime number or not**


```python
import math
num = int(input("Enter an integer: "))

# Check if the number is prime
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            is_prime = False
            break

# Print the result
if is_prime:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")


```

16. # **Python program to generate the prime numbers from 1 to 20**


```python
# Generate prime numbers from 1 to 20
primes = []
for num in range(1, 21):
    if num > 1:
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)

# Print the prime numbers
print("Prime numbers from 1 to 20:")
for prime in primes:
    print(prime)
```

17. # **Python program to find the roots of a quadratic equation**


```python
import math

a = float(input("Enter the coefficient of x^2: "))
b = float(input("Enter the coefficient of x: "))
c = float(input("Enter the constant term: "))

# Calculate the discriminant
discriminant = b**2 - 4*a*c

# Check the nature of the roots
if discriminant > 0:
    # Two distinct real roots
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print("The equation has two distinct real roots:")
    print("Root 1:", root1)
    print("Root 2:", root2)
elif discriminant == 0:
    # One real root (repeated)
    root = -b / (2*a)
    print("The equation has one real root (repeated):")
    print("Root:", root)
else:
    # Complex roots
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(-discriminant) / (2*a)
    print("The equation has complex roots:")
    print("Root 1:", real_part, "+", imaginary_part, "i")
    print("Root 2:", real_part, "-", imaginary_part, "i")

```

21. # **Python program to find the odd numbers in an array**

22. # **Python program to find the largest number in a list without using built-in functions**


```python

numbers = input("Enter the numbers (separated by spaces): ").split()
numbers = [int(num) for num in numbers]

# Initialize a variable to store the largest number
largest = numbers[0]

# Iterate through the numbers and update the largest number if necessary
for num in numbers:
    if num > largest:
        largest = num

# Print the largest number
print("The largest number in the list is:", largest)
```

23. # **Python program to delete an element from a list by index**


```python

numbers = input("Enter the numbers in the list (separated by spaces): ").split()
numbers = [int(num) for num in numbers]
index = int(input("Enter the index of the element to delete: "))

# Delete the element at the specified index
if index < 0 or index >= len(numbers):
    print("Invalid index!")
else:
    del numbers[index]
    print("Element at index", index, "deleted.")

# Print the updated list
print("Updated list:", numbers)
```

28. # **Python program to check leap year**


```python

year = int(input("Enter a year: "))

# Checking if the year is a leap year
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
